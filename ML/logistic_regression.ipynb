```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support
```

```python
data = pd.read_csv("./Titaanic Datasetforlogisticregression .csv")
data.head()
```

```python
# 결측값 확인
data.isna().sum()
```

```python
data.info()
```

```python
# 라벨 인코딩
def label_encoder(df: pd.DataFrame) -> pd.DataFrame:
  for col in df.columns:
      label = {}
      if df[col].dtype == "object":
        for i, value in enumerate(df[col].unique()):
          label[value] = i
        df[col] = df[col].map(lambda x: label[x])
  return df

data = label_encoder(data)
data.head()
```

```python
X_data = data.drop(['Survived'], axis=1)
target = data['Survived']
```

```python
# 데이터 정규화
mms = MinMaxScaler()
cols = ["Age", "Siblings/Spouses Aboard", "Parents/Children Aboard", "Fare"]
X_data[cols] = mms.fit_transform(X_data[cols])
X_data.head()
```

```python
# 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X_data, target)
```

```python
class LogisticRegression():
  def __init__(self, learning_rate=0.01, epoch=100000):
    self._learning_rate = learning_rate
    self._epoch = epoch

  # sigmoid func
  def sigmoid(self, z):
      return 1.0 / (1.0 + np.exp(-z))

  # cross entropy func
  def loss(self, h, y):
    delta = 1e-7
    return -np.sum(y * np.log(h + delta)) / len(h)

  def fit(self, X, y):
    sample_len, feature_len = np.shape(X)

    self._W = np.zeros(feature_len)

    for i in tqdm(range(self._epoch)):
      z = np.dot(X, self._W)
      h = self.sigmoid(z)

      residual = h - y

      loss = self.loss(h, y)
      gradient = np.dot(X.transpose(), residual) / sample_len

      self._W -= gradient * self._learning_rate

  def predict_proba(self, X):
    return self.sigmoid(np.dot(X, self._W))

  def predict(self, X):
    return self.predict_proba(X).round()
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict_proba(X_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
# 모델 평가
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# 정확도
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# 정밀도, 재현율, F-score
precision_recall_fscore_support(y_test, y_pred, average='binary')
```
